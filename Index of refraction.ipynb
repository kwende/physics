# imports and stuff

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define constants and functions

```python
velocity_in_substance1 = 10
velocity_in_substance2 = 5

pos_of_start = (-100, -50)
pos_of_end = (100, 50)

def distance(p1, p2):
    return np.linalg.norm(np.array(p1) - np.array(p2))
```

# test distance between points (straight line)

```python
distance(pos_of_start, pos_of_end)
```

```python
distances = []
times = []

for i in range(-100, 100, 5):
    point_on_beach = (i, 0)
    d1 = distance(pos_of_start, point_on_beach)
    d2 = distance(point_on_beach, pos_of_end)

    t1 = 1/(velocity_in_substance1 / d1)
    t2 = 1/(velocity_in_substance2 / d2)

    times.append(t1 + t2)
    distances.append(d1 + d2)
```

# show distances are parabolic (there is a global minimum)

```python
x = np.arange(0, len(distances))
plt.plot(x, distances)
```

# show that time also have a global minimum in the shape of the Feynman lectures

```python
plt.plot(x, times)
```
